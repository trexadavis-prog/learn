# Module 4: Functions & Modular Code (Concept Guide)

Welcome to Module 4! As you write larger programs, you don't want to type the same logic over and over. 

We use **Functions** to package a block of code with a name, so we can run it whenever we want by simply calling its name. Think of a function like a **blender**: you throw ingredients in (inputs), press a button (run code), and get a smoothie out (output).

In this guide, we will learn:
1. How to create and run functions (`def` and `return`).
2. How to pass information into functions (**Parameters**).
3. How to use default fallback values (**Default Parameters**).
4. Where variables live (**Local vs. Global Scope**).
5. How to import code from Python's libraries (`import`).

## 1. Defining and Calling Functions

- We define a function using the **`def`** keyword, followed by the function name, parentheses `()`, and a colon `:`.
- The code block inside the function must be indented.
- We use **`return`** to send the output back. If a function doesn't return anything, it returns `None` (empty) by default.

```python
# Defining a function that doubles a number
def double_number(x):
 result = x * 2
 return result

# Calling the function (using its name and passing an argument)
output_value = double_number(5)
print("Doubling 5 gives:", output_value)
print("Doubling 100 gives:", double_number(100))
```

## 2. Parameters & Default Values

- **Parameters:** The variables inside the function definition parentheses (e.g. `x` in `double_number(x)`). They are placeholders for inputs.
- **Arguments:** The actual values you pass when calling it (e.g. `5` or `100`).
- **Default parameters:** Fallback values used if the caller doesn't provide an argument.

```python
# A function with default parameters
def greet(name="Friend", greeting="Hi"):
 return f"{greeting}, {name}!"

print(greet("Alice", "Good morning")) # Uses custom arguments
print(greet("Bob")) # Uses default greeting ('Hi')
print(greet()) # Uses both defaults ('Hi, Friend!')

# Using keyword arguments (naming the parameters explicitly)
print(greet(greeting="Howdy", name="Charlie"))
```

## 3. Variable Scope (Local vs. Global)

Variables are not accessible everywhere in your code. Where they live is called their **Scope**:

- **Local variables:** Variables created *inside* a function. They only exist inside that function. You cannot use or print them outside the function.
- **Global variables:** Variables created *outside* functions. They are accessible anywhere.

```python
company_name = "Python Corp" # Global variable - readable anywhere

def show_stats():
 employees = 50 # Local variable - only exists inside this function
 print("Inside function:", company_name)
 print("Inside function:", employees)

show_stats()

print("Outside function:", company_name)

# Trying to access 'employees' here will crash because it's local to show_stats():
try:
 print(employees)
except NameError as e:
 print("Error:", e)
```

## 4. Importing Modules (Built-in Toolboxes)

Python has a massive collection of pre-made code modules you can use for free! You load them using `import`.

Two common built-in modules are:
- **`math`:** Includes constants like Pi and functions like square roots.
- **`random`:** Used for rolling dice, picking items from a list, or generating random numbers.

```python
import math
import random

print("Square root of 81:", math.sqrt(81)) # 9.0

# Roll a random 6-sided die
die_roll = random.randint(1, 6)
print("You rolled a:", die_roll)

# Pick a random color
colors = ["red", "green", "blue"]
print("Picked color:", random.choice(colors))
```
