## 1. Making Python Speak & Writing Comments

To display information on the screen, we use a tool called `print()`. Think of it as a megaphone. You put whatever text you want Python to say inside quotes `"` and parentheses `()`.

### What are Comments?
Comments start with a `#` symbol. Python completely ignores everything written after a `#` on that line. They are notes written for humans to explain what is happening.

````python
# This is a comment. Python will skip this line.
print("Hello! Welcome to Python!") # This makes Python display the message on the screen.
````

## 2. Variables: Labeled Storage Boxes

Imagine you are packing for a move. You put an object inside a box and write a label on the outside so you know what is inside. 

In programming, a **variable** is exactly like a labeled box. 
- The **label** is the variable name (e.g., `user_name` or `lucky_number`).
- The **contents** is the value inside (e.g., `"Alice"` or `7`).

To store something in a box, we use the `=` sign (assignment operator):
```python
box_name = value_to_store
```

````python
# Storing a string (text) in a box labeled 'favorite_animal'
favorite_animal = "Panda"

# Storing a number in a box labeled 'number_of_pets'
number_of_pets = 2

# To use the boxes, we just call them by their labels!
print(favorite_animal)
print(number_of_pets)
````

### Variable Naming Rules:
1. **Start with a letter or underscore** (like `my_var` or `_var`). You cannot start with a number (like `1var`).
2. **Only use letters, numbers, and underscores**. No spaces or punctuation. For multiple words, use underscores (called snake_case: `my_dog_name`).
3. **Capitalization matters!** `Age`, `age`, and `AGE` are three different variables.

## 3. Different Types of Information (Data Types)

Python categorizes everything we store. Here are the 4 main types of data you will use most:

- **String (`str`):** Text. Always surrounded by quotes `"..."` or `'...'` (e.g., `"Hello!"`).
- **Integer (`int`):** Whole numbers without a decimal point (e.g., `10`, `-3`).
- **Float (`float`):** Decimal numbers (e.g., `3.14`, `10.0`).
- **Boolean (`bool`):** Logical true/false. Can only be `True` or `False` (note the capital letters).

You can inspect what type of data is in a variable using `type(variable_name)`.

````python
greeting = "Good morning!"
cats = 3
price = 19.99
is_raining = False

print(type(greeting)) # Output: <class 'str'>
print(type(cats)) # Output: <class 'int'>
print(type(price)) # Output: <class 'float'>
print(type(is_raining)) # Output: <class 'bool'>
````

## 4. Basic Math (Arithmetic)

Python can be used as a simple calculator. Here are the symbols you use:

- `+` for addition
- `-` for subtraction
- `*` for multiplication
- `/` for normal division (always results in a decimal, e.g. `5 / 2 = 2.5`)
- `//` for integer division (throws away the decimal part, e.g. `5 // 2 = 2`)
- `%` for modulo (gives you *only* the remainder, e.g. `5 % 2 = 1`)
- `**` for exponents/powers (e.g. `5 ** 2 = 25`)

````python
cookies = 10
friends = 3

cookies_per_friend = cookies // friends # How many whole cookies each get
leftover_cookies = cookies % friends # The remaining cookies that can't be shared evenly

print("Cookies per friend:", cookies_per_friend)
print("Leftovers:", leftover_cookies)
````

## 5. String Formatting (F-Strings)

What if we want to print a sentence that includes values from our variables? 
The modern, easiest way is to use **F-strings** (Formatted Strings). You simply put the letter `f` right before the first quote, and then write your variables inside curly braces `{}`. Python will swap out the curly braces for the actual value.

````python
name = "Bob"
score = 95

# Note the f before the quote!
message = f"Congratulations {name}! Your score is {score}%."
print(message)
````

## 6. Asking the User for Information (`input`)

You can make your programs interactive by asking the user to type something using the `input()` function.

> [!WARNING]
> **Crucial Rule:** The `input()` function **always** reads the user's input as text (a String), even if they typed a number! 
>
> If you need to perform calculations with a number the user typed, you must convert the text to a number first using `int()` (whole numbers) or `float()` (decimal numbers).

````python
# Reading name (which is naturally text)
user_name = input("What is your name? ")
print(f"Hello, {user_name}!")

# Reading age (which is a number we might want to do math on)
age_text = input("How old are you? ")
age_number = int(age_text) # Convert the text "25" to the integer number 25

age_next_year = age_number + 1
print(f"{user_name}, next year you will be {age_next_year} years old!")
````
