4. If the player guesses the number, they win! If they run out of attempts, they lose.

secret_number = random.randint(1, 50)

while attempts_left > 0:
 # Get input
 guess = int(input(f"Attempts left: {attempts_left}. Enter your guess: "))
 attempts_left = attempts_left - 1

 print(f" Congratulations! You guessed the secret number {secret_number}!")
 won = True
 break
 elif guess < secret_number:
 print("Too low! Try a higher number.")
 else:
 print("Too high! Try a lower number.")
        
if not won:
