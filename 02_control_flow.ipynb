# Module 2: Control Flow & Logic (Concept Guide)

Welcome to Module 2! So far, your Python code ran straight from top to bottom. But real programs need to make decisions and repeat actions. This is called **Control Flow**.

In this guide, we will learn:
1. How to compare values (**Comparison Operators**).
2. How to make decisions in code (**If, Elif, and Else Statements**).
3. How to combine conditions (**Logical Operators**).
4. How to repeat tasks (**For and While Loops**).
5. How to stop or skip inside loops (`break` and `continue`).

## 1. Comparing Values (Comparison Operators)

Comparison operators ask a question. The result is always a Boolean: either `True` (yes) or `False` (no).

Here are the questions you can ask Python:

- `==` : Are these values **equal**? *(Note: double equals `==` checks equality, single equals `=` assigns variables)*
- `!=` : Are these values **not equal**?
- `>` : Is the left value **greater than** the right value?
- `<` : Is the left value **less than** the right value?
- `>=` : Is the left value **greater than or equal to** the right value?
- `<=` : Is the left value **less than or equal to** the right value?

````python
cookies_in_jar = 5
max_cookies_allowed = 3

print("Are they equal?", cookies_in_jar == max_cookies_allowed) # False
print("Do we have too many?", cookies_in_jar > max_cookies_allowed) # True
````

## 2. Decision Making (`if`, `elif`, `else`)

Imagine a fork in the road. Depending on a condition, you take one path or another. 

We write this in Python using `if`, `elif` (short for "else if"), and `else`:

```python
if condition_is_true:
 # do this
elif another_condition_is_true:
 # do this instead
else:
 # do this if none of the above are true
```

> [!IMPORTANT]
> **The Indentation Rule:** Notice the spaces before the code inside the block. In Python, you must indent (usually 4 spaces or pressing Tab) the lines under an `if`, `elif`, or `else`. This tells Python: "These lines belong inside this decision block."

````python
temperature = 15 # degrees Celsius

if temperature > 25:
 print("It's hot outside! Wear a t-shirt. ")
elif temperature > 10:
 print("It's mild. A sweater is perfect. ") # This will run!
else:
 print("It's freezing! Wear a heavy coat. ")
````

## 3. Combining Conditions (Logical Operators)

Sometimes, a decision depends on multiple things. We combine conditions using:
- `and`: `True` only if **both** parts are True.
- `or`: `True` if **at least one** part is True.
- `not`: Flips the result (e.g. `not True` becomes `False`).

````python
has_ticket = True
knows_manager = False

if has_ticket or knows_manager:
 print("You can enter the concert! ") # Runs because at least one is True

is_weekend = True
is_raining = True
if is_weekend and not is_raining:
 print("Let's go for a picnic!")
else:
 print("Better stay indoors. ")
````

## 4. Loops: Repeating Actions

Instead of writing the same line of code 10 times, we use loops to repeat tasks.

### The `for` Loop (Counting / Iterating)
We use a `for` loop when we know how many times we want to repeat a task, or when we want to go through a sequence of items. 
- The `range(start, end)` function generates numbers starting from `start` up to, but not including, `end`.

````python
# Counting from 1 to 3
for count in range(1, 4):
 print(f"This is loop number {count}")
````

### The `while` Loop (Until a Condition Changes)
We use a `while` loop when we don't know exactly how many repetitions we need, but we want to keep going **while a condition remains True**.

> [!WARNING]
> **Infinite Loops:** If the condition remains `True` forever, your program will freeze because it keeps running. Make sure you change the condition inside the loop (e.g. subtracting 1 from a counter) so it eventually stops!

````python
oil_level = 3

while oil_level > 0:
 print(f"Engine running. Oil level: {oil_level}")
 oil_level = oil_level - 1 # Decrease level so loop can stop

print("Out of oil! Engine stopped.")
````

### Loop Control: Break & Continue
- `break` tells Python: "Stop the loop immediately and get out!"
- `continue` tells Python: "Skip the rest of this turn and go straight to the next loop iteration."

````python
print("--- Loop with break (stops at 3) ---")
for num in range(1, 6):
 if num == 3:
 break
 print(num)

print("\n--- Loop with continue (skips 3) ---")
for num in range(1, 6):
 if num == 3:
 continue
 print(num)
````
