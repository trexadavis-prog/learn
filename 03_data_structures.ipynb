# Module 3: Data Structures (Concept Guide)

Welcome to Module 3! So far, we've only stored single values in our variables (like `x = 5` or `name = "Alice"`). 

But what if you need to store a list of 100 groceries, or a phone directory? We use **Data Structures** to hold groups of values. 

In this guide, we will learn Python's 4 main data structures:
1. **Lists:** Ordered sequences of items (like a grocery list).
2. **Dictionaries:** Key-value mappings (like a contacts book or dictionary).
3. **Tuples:** Locked, unchangeable sequences.
4. **Sets:** Collections of unique items (no duplicates allowed).

## 1. Lists: The Grocery List

A **list** is an ordered group of items. We write a list inside square brackets `[]` with commas in between.

```python
shopping_list = ["apples", "milk", "bread"]
```

### Indexing & Slicing (Finding Items)
Python orders lists by numbers starting at **`0`** (this is called the index). 
- To get the first item, we use index `0`: `list[0]`.
- To get items from the end, we count backward using negative numbers (e.g., `list[-1]` is the last item).
- To get a subset, we use slicing `list[start:stop]` (which grabs items from `start` up to but not including `stop`).

````python
animals = ["cat", "dog", "bird", "rabbit"]

print("First animal:", animals[0]) # cat
print("Last animal:", animals[-1]) # rabbit
print("Middle slice:", animals[1:3]) # dog, bird
````

### Modifying Lists
Lists are mutable (they can be changed). Here is how you modify them:
- **Add item to end:** `list.append(new_item)`
- **Insert item at a position:** `list.insert(index, new_item)`
- **Remove item by name:** `list.remove(item_to_remove)`
- **Get length:** `len(list)`

````python
playlist = ["Song A", "Song B"]
playlist.append("Song C") # Adds "Song C" to the end
playlist.insert(1, "Bonus Track") # Inserts at index 1 (pushes B & C down)
playlist.remove("Song A") # Removes "Song A"

print("Playlist:", playlist)
print("Total songs:", len(playlist))
````

## 2. Dictionaries: The Contact Book

A **dictionary** stores pairs of information. Instead of using a number to find something, you use a custom label called a **Key** to look up a **Value**.

We use curly braces `{}` and separate keys and values with colons `key: value`.

```python
phonebook = {"Alice": "555-0199", "Bob": "555-0210"}
```

````python
# Creating a dictionary
car = {
 "brand": "Ford",
 "model": "Mustang",
 "year": 1964
}

# Accessing values by their key
print("Car Brand:", car["brand"])

# Adding or modifying pairs
car["year"] = 2023 # Modifies the value for 'year'
car["color"] = "Red" # Adds a new key-value pair

print("Car Info:", car)
````

### Helper Dictionary Methods
- Get all keys: `dict.keys()`
- Get all values: `dict.values()`
- Loop through keys and values: `for key, val in dict.items():`

````python
prices = {"milk": 2.50, "bread": 1.80}

for item, price in prices.items():
 print(f"An order of {item} costs ${price:.2f}")
````

## 3. Tuples: Locked lists

A **tuple** is an ordered list that **cannot be changed** once created. We write tuples inside parentheses `()`.

Use tuples when you want to protect data from being accidentally modified (like coordinates or RGB values).

````python
screen_resolution = (1920, 1080)
print("Width:", screen_resolution[0])

# Running this will cause an error because tuples are locked (immutable):
try:
 screen_resolution[0] = 2560
except TypeError as e:
 print("Error:", e)
````

## 4. Sets: Collections of Unique Elements

A **set** is an unordered collection of items where **every item must be unique**. We write sets using curly braces `{}` containing values directly (no keys).

Sets automatically filter out duplicates. They are great for checking if something is in a group or performing mathematical comparisons (like finding common elements between lists).

````python
guest_list = {"Alice", "Bob", "Alice", "Charlie"} # Note the duplicate 'Alice'
print("Unique guests:", guest_list) # Duplicates are auto-removed!

# Checking if someone is in the set (very fast!)
print("Is Bob invited?", "Bob" in guest_list)
````
